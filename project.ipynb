### Joining all datasets

```python
df = acquisitions.merge(degrees,on = 'id', suffixes = ['_acquisitions','_degrees'])\
    .merge(funding_rounds, on = 'id' , suffixes = ['','_funding_rounds'])\
    .merge(funds, on = 'id' , suffixes = ['','_funds'])\
    .merge(investments, on = 'id' , suffixes = ['','_investments'])\
    .merge(ipos, on = 'id' , suffixes = ['','_ipos'])\
    .merge(milestones, on = 'id' , suffixes = ['','_milestones'])\
    .merge(objects, on = 'id' , suffixes = ['','_objects'])\
    .merge(offices, on = 'id' , suffixes = ['','_offices'])\
    .merge(people, on = 'id' , suffixes = ['','_people'])\
    .merge(relationships, on = 'id' , suffixes = ['','_relationships'])
```
